```python
salary = pd.read_csv('Salary Data.csv')
salary.head()
```

```python
plt.figure()
sns.scatterplot(x= 'Age', y='Salary', data=salary)
plt.title('Default Scatter Plot: Age Vs Salary')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x= 'Age', y='Salary', hue= 'Education Level', style= 'Gender', data=salary)
plt.title('Age Vs Salary by education level and gender')
plt.show()
```

```python
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
plt.figure()
sns.scatterplot(x= 'Age', y='Salary', hue= 'Education Level', style= 'Gender', data=salary, palette= education_colors)
plt.title('Age Vs Salary by education level and gender')
plt.show()
```

```python
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers= {"Female": "o", "Male": "s"}
plt.figure()
sns.scatterplot(x= 'Age', y='Salary', hue= 'Education Level', style= 'Gender', data=salary, palette= education_colors)
plt.title('Age Vs Salary by education level and gender')
plt.show()
```

```python
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers= {"Female": "o", "Male": "s"}
plt.figure(figsize=(12,8))
sns.set_style("ticks")
sns.scatterplot(x= 'Age', y='Salary', hue= 'Education Level', style= 'Gender', data=salary, palette= education_colors)
plt.title('Age Vs Salary by education level and gender')
plt.show()
```

```python
education_colors= {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers= {"Female": "o", "Male": "s"}
plt.figure(figsize=(12,8))
sns.set_style("ticks")
sns.scatterplot(x= 'Age', y='Salary', hue= 'Education Level', style= 'Gender', data=salary, palette= education_colors)
plt.legend(ncol=2, frameon= False, loc="lower right")
plt.title('Age Vs Salary by education level and gender')
plt.show()
```

```python
gender_colors= {"Female": "green", "Male": "blue"}
plt.figure(figsize=(12,8))
sns.set_style("ticks")
Bachelors=sns.scatterplot(x="Age", y="Salary", hue= "Gender", data= salary[salary["Education Level"]== "Bachelor's"],palette= gender_colors)
Bachelors.set_xlim(25,50)
Bachelors.set_ylim(50000,250000)
plt.legend(ncol=2, frameon= False, loc="lower right")
plt.title('Age Vs Salary by education level and gender')
plt.show()
```

```python
# Creating a separate scatter plot for each Education Level for better clarity in visualization -MASTERS
#Since we won't have a hue for Education level, we will define a custom pallete for the gender now
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks") #no gridlines
Masters=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=salary[salary["Education Level"] == "Master's"], palette=gender_colors)

# Setting limits for x and y axes
Masters.set_xlim(25, 50)  
Masters.set_ylim(50000, 250000)
#customazing the legend
plt.legend(frameon=False, loc= "upper left" )
plt.title("Age vs Salary Based on Gender Among Master's Degree Holders")
plt.show()
```

```python
## Creating a FacetGrid with subplots for each education level
#defining the palette for the gender
gender_colors = {"Female": "red", "Male": "blue"}

g = sns.FacetGrid(salary, col="Education Level")
g.map_dataframe(sns.scatterplot, x= "Age", y= "Salary", hue="Gender", palette=gender_colors)  
#Adjusting the legend
g.add_legend(title="Gender", frameon=False, loc="lower right")

plt.show()
```

```python
#Creating a figure with 3 axis
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
gender_colors = {"Female": "red", "Male": "blue"}
sns.set_style("ticks") #no gridlines
# Use Seaborn's scatterplot for the fourth subplot
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=salary[salary["Education Level"] == "Bachelor's"], palette=gender_colors, ax=ax[0])
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=salary[salary["Education Level"] == "Master's"], palette=gender_colors, ax=ax[1])
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=salary[salary["Education Level"] == "PhD"], palette=gender_colors, ax=ax[2])

for axes in ax:
    axes.legend(frameon=False)

# Set the title in the center
plt.suptitle('Scatter Plot: Age vs Salary by Education Level and Gender', ha="center")

# Show the plot
plt.show()
```

```python
#Creating a figure with 3 axis and setting the limits
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
gender_colors = {"Female": "red", "Male": "blue"}
sns.set_style("ticks") #no gridlines
scatter_1 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=salary[salary["Education Level"] == "Bachelor's"],
                            palette=gender_colors, ax=ax[0])
scatter_2 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=salary[salary["Education Level"] == "Master's"],
                            palette=gender_colors, ax=ax[1])
scatter_3 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=salary[salary["Education Level"] == "PhD"],
                            palette=gender_colors, ax=ax[2])

plt.suptitle('Scatter Plot: Age vs Salary by Education Level and Gender', ha="center")
# Setting the limits for x and y axes
for axes in ax:
    axes.set_xlim(25, 50)  
    axes.set_ylim(50000, 250000) 
    axes.legend(frameon=False, loc = "upper left")

# Show the plot
plt.show()
```
